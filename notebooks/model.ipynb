```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

from rdkit import Chem
from rdkit.Chem import Descriptors
```

```python
df = pd.read_csv('df3.csv')
```

```python
df
```

```python
Y = df['Resistance phenotype'].dropna()
```

```python
x = df.drop(columns = 'Resistance phenotype')
```

```python
x
```

```python
y = df['Resistance phenotype']
```

```python
y.dropna(subset=['Resistance phenotype'], inplace=True)
```

```python
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python
#label encoding antibiotics
if 'antibiotics' in df.columns:  
    x['antibiotics'] = label_encoder.fit_transform(x['antibiotics'])
```

```python
if 'Isolation source' in df.columns:
    x['Isolation source'] = label_encoder.fit_transform(x['Isolation source'])
```

```python
x
```

```python
def smiles_to_features(smiles):
    mol = Chem.MolFromSmiles(smiles)
    if mol is not None:
        # Calculate various descriptors
        return [
            Descriptors.NumHDonors(mol),          
            Descriptors.NumHAcceptors(mol),      
            Descriptors.NumRotatableBonds(mol),  
            Descriptors.TPSA(mol),                              
        ]
    else:
        return [0]*4
```

```python
features = df['CanonicalSMILES'].apply(smiles_to_features)
```

```python
features
```

```python
features_df = pd.DataFrame(features.tolist(), columns=['NumHDonors', 'NumHAcceptors', 
    'NumRotatableBonds', 'TPSA'])
```

```python
features_df
```

```python
X = pd.concat([features_df, x], axis=1)
```

```python
X = X.drop(columns = 'CanonicalSMILES')
```

```python
X
```

```python
y
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
print(classification_report(y_test, y_pred, target_names=['susceptible', 'resistant','unknown']))
```
